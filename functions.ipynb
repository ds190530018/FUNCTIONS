```python
#QUS8.  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(1, 3), (4, 1), (2, 5), (3, 2)]
sorted_data = sorted(data,key=lambda x:x[1])
print(sorted_data)
```

```python
#QUS.9  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius_temperatures = [0,10,20,30,40]
fahrenheit_tempratures = list(map(lambda x:(x*9/5)+32,celsius_temperatures))
print(fahrenheit_tempratures)

```

```python
#QUS10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(input_string):
  vowels = "aeiousAEIOU"
  return "".join(filter(lambda x:x not in vowels,input_string))
```
